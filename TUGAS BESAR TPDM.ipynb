```python
x =df.iloc[0:,4:5]
x
```

```python
kmeans = KMeans(4)
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters = df.copy()
data_with_clusters['Clusters'] = identified_clusters 
plt.scatter(data_with_clusters['nama kabupaten kota'],data_with_clusters['jenis tempat hiburan'],c=data_with_clusters['Clusters'],cmap='rainbow')
```

```python
wcss=[]
for i in range(1,10):
    kmeans = KMeans(i)
kmeans.fit(x)
wcss_iter = kmeans.inertia_
wcss.append(wcss_iter)

number_clusters = range(10,11)
plt.plot(number_clusters,wcss)
plt.title('The Elbow title')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```
